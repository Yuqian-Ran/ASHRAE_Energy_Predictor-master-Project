```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import warnings
```

```python
def time_format(df):
    try:
        df["timestamp"] = pd.to_datetime(df["timestamp"], format='%Y-%m-%d %H:%M:%S')
    except:
        print('There is no timestamp column')
    return df
```

```python
# define a making pivot figure function
def pivot_figure(df_pivot_2016, df_pivot_2017, df_pivot_2018, max_id, name_list):
    warnings.filterwarnings("ignore")
    # check 10 days data
    start_time = datetime.date(2016, 1, 1)
    end_time = datetime.date(2016, 1, 10)
    # make figures
    plt.figure(figsize = (18,6))
    plt.plot(df_pivot_2016.loc[start_time:end_time, max_id], markerfacecolor='blue', label = name_list[0])
    plt.plot(df_pivot_2017.loc[start_time:end_time, max_id], markerfacecolor='yellow', label = name_list[1])  
    plt.plot(df_pivot_2018.loc[start_time:end_time, max_id], markerfacecolor='olive', label = name_list[2])
    plt.title(name_list[3])
    plt.savefig("./Output/png/" + name_list[4], bbox_inches = 'tight', dpi = 300)
    plt.legend()
    plt.show()
```

```python
# import csv files
df_weather_train = pd.read_csv('../Resources/weather_train.csv')
df_weather_test = pd.read_csv('../Resources/weather_test.csv')
# change the format of timestamp
df_weather_train = time_format(df_weather_train)
df_weather_test = time_format(df_weather_test)
# make the pivot for dataframe to make the figure
pivot_weather_train = df_weather_train.pivot(index = 'timestamp', columns = 'site_id', values = 'air_temperature')
# find the max nan value site id
max_id = pivot_weather_train.count().idxmin()
# site all weather data to same time by minus 1 year
df_weather_test['timestamp'] = df_weather_test['timestamp'] - datetime.timedelta(366)
pivot_weather_2017 = df_weather_test.pivot(index = 'timestamp', columns = 'site_id', values = 'air_temperature')
df_weather_test['timestamp'] = df_weather_test['timestamp'] - datetime.timedelta(365)
pivot_weather_2018 = df_weather_test.pivot(index = 'timestamp', columns = 'site_id', values = 'air_temperature')
name_list = ['2016', '2017', '2018', 'site 15 air_temperature initial data', 'site_15_air_temp_initial.png']
pivot_figure(pivot_weather_train, pivot_weather_2017, pivot_weather_2018, max_id, name_list)
```

```python
def fill_poly(df_pivot):
    return df_pivot.fillna(df_pivot.interpolate(method='polynomial', order=3))
def fill_linear(df_pivot):
    return df_pivot.fillna(df_pivot.interpolate(method='linear'))
def fill_both(df_pivot):
    df_pivot = (df_pivot.fillna(df_pivot.interpolate(method='linear', limit_direction='both')) +
                df_pivot.fillna(df_pivot.interpolate(method='polynomial', order=3, limit_direction='both'))) * 0.5
    # workaround: fill last NANs with neighbour
    # only the first item is missing
    assert df_pivot.count().min() >= len(df_pivot)-1
    # fill with second item
    return df_pivot.fillna(df_pivot.iloc[1])         
```

```python
# plot 2 different method of interpolation of polynomial and linear
# reload csv file
df_weather_test = pd.read_csv('../Resources/weather_test.csv')
df_weather_test = time_format(df_weather_test)
pivot_weather_train = df_weather_train.pivot(index = 'timestamp', columns = 'site_id', values = 'air_temperature')
# find the max nan value site id
max_id = pivot_weather_train.count().idxmin()
pivot_poly = fill_poly(pivot_weather_train)
pivot_linear = fill_linear(pivot_weather_train)
pivot_both = fill_both(pivot_weather_train)
name_list = ['polynomial', 'linear', 'both', 'site 15 air_temperature fill data', 'site_15_air_temp_filling_test.png']
pivot_figure(pivot_poly, pivot_linear, pivot_both, max_id, name_list)
# Cubic interpolation is generally the best option but the data is a bit too noisy for it, 
# so 50/50% mix with linear interpolation seems better
```

```python
def real_fill_nan(df_weather):
    df = None
    for col in ['air_temperature', 'dew_temperature']:
        df_temp = fill_both(df_weather.pivot(index='timestamp', columns='site_id', values=col))
        df_temp = df_temp.sort_index().unstack().to_frame(col)
        if df is None:
            df = df_temp
        else:
            df[col] = df_temp[col]
    return df
```

```python
# fill the nan by mixture interpolation and change negative value to 0 in 
# 'cloud_coverage', 'precip_depth_1_hr', 'wind_direction', 'wind_speed'
for i in ['train', 'test']:
    df_weather = pd.read_csv(f'../Resources/weather_{i}.csv')
    df_weather = time_format(df_weather)
    df_temp = real_fill_nan(df_weather)
    columns_name = ['site_id', 'timestamp', 'cloud_coverage', 'precip_depth_1_hr', 'wind_direction', 'wind_speed']
    df_temp = df_temp.reset_index().merge(df_weather[columns_name], how='left', on=['site_id', 'timestamp'])\
              .set_index(['site_id', 'timestamp'])
    for j in ['cloud_coverage', 'precip_depth_1_hr', 'wind_direction', 'wind_speed']:
        df_temp.loc[df_temp[j] < 0, j] = 0
        df_temp.fillna(0, inplace=True)
    df_temp.reset_index().to_csv(f'./Output/csv/weather_{i}_cleaned.csv', index = False)
```
